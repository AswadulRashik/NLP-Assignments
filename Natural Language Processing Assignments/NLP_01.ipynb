```python
import pandas as pd


given_file = "/content/IMDB Dataset.csv"
data = pd.read_csv(given_file)

print(f"Total number of reviews: {len(data)}")

```

```python
from nltk.tokenize import sent_tokenize, word_tokenize


data['word_count'] = data['review'].apply(lambda review: len(word_tokenize(review)))
data['sentence_count'] = data['review'].apply(lambda review: len(sent_tokenize(review)))

most_words_idx = data['word_count'].idxmax()
most_words_row = data.loc[most_words_idx]

print("\nRow Index with Most Words:")
print(f"Review: {most_words_row['review']}")
print(f"Row with Most Words: {most_words_idx}")
print(f"Word Count: {most_words_row['word_count']}")


most_sentences_idx = data['sentence_count'].idxmax()
most_sentences_row = data.loc[most_sentences_idx]

print("\nRow Index with Most Sentences:")
print(f"Review: {most_sentences_row['review']}")
print(f"Row with most sentences: {most_sentences_idx}")
print(f"Sentence Count: {most_sentences_row['sentence_count']}")

```

```python
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)


def clean_review(review):
    tokens = word_tokenize(review)

    tokens = [word for word in tokens if word.lower() not in stop_words and word not in punctuations]
    return ' '.join(tokens)

data['clean_review'] = data['review'].apply(clean_review)
```

```python
data['clean_review'] = data['clean_review'].str.lower()

with open("clean_reviews.txt", "w", encoding="utf-8") as file:
    file.writelines(review + '\n' for review in data['clean_review'])

with open("clean_reviews.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
print(f"Number of saved reviews: {len(lines)}")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
given_file = "/content/IMDB Dataset.csv"
df = pd.read_csv(given_file)


stop_words = set(stopwords.words("english"))


def clean_review(review):
    review = review.lower()
    words = word_tokenize(review)


    words_no_stopwords = [w for w in words if w not in stop_words]
    cleaned_words = [w for w in words_no_stopwords if w.isalpha()]


    cleaned_review = " ".join(cleaned_words)
    return cleaned_review


df["cleaned_review"] = df["review"].apply(clean_review)


print(df.head())
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["cleaned_review"])


print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
print("Sample features (words):", vectorizer.get_feature_names_out()[:10])
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


def load_glove_embeddings(file_path):
    embeddings = {}
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.array(values[1:], dtype="float32")
            embeddings[word] = vector
    return embeddings


glove_path = "/content/glove.6B.100d.txt"
embeddings = load_glove_embeddings(glove_path)


def calculate_cosine_similarity(vec1, vec2):
    return cosine_similarity([vec1], [vec2])[0][0]


word_pairs = [("man", "woman"), ("cat", "dog"), ("king", "queen")]
for w1, w2 in word_pairs:
    similarity = calculate_cosine_similarity(embeddings[w1], embeddings[w2])
    print(f"Cosine similarity between {w1} and {w2}: {similarity:.4f}")


king_vector = embeddings["king"]
man_vector = embeddings["man"]
woman_vector = embeddings["woman"]
queen_vector = embeddings["queen"]


result_vector = king_vector - man_vector + woman_vector


similarity_to_queen = calculate_cosine_similarity(result_vector, queen_vector)
print(f"Cosine similarity between 'King - Man + Woman' and 'Queen': {similarity_to_queen:.4f}")
```
